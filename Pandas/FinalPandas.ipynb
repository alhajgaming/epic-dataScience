```python
import pandas as pd
import numpy as np

data = {
    "Name":       ["Rahul", "Priya", None,    "Sneha", "Karan"],
    "Age":        [23,       27,      25,       None,    30],
    "Salary":     [45000,    None,    58000,    40000,   None],
    "Department": ["IT",     "HR",    "IT",     None,    "Sales"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Count missing values per column
print(df.isnull().sum())
# See missing values as True/False
print(df.isnull())

# Percentage of missing values — very useful!
print(df.isnull().sum() / len(df) * 100)
```

```python
# Drop ALL rows that have ANY missing value
clean_df = df.dropna()
print(clean_df)
```

```python
# Drop rows only if ALL values are missing
clean_df = df.dropna(how="all")
clean_df = df.dropna(axis=1)
```

```python
print(clean_df)
```

```python
# Fill missing Salary with the average salary
avg_salary = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(avg_salary)
print(df["Salary"])
```

```python
# Fill missing Department with "Unknown"
df["Department"] = df["Department"].fillna("Unknown")

# Fill missing Age with median age
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Name with "No Name"
df["Name"] = df["Name"].fillna("No Name")
print(df)
```

```python
data = {
    "Name":   ["Rahul", "Priya", "Arjun"],
    "Salary": ["45000", "62000", "58000"],  # ← stored as text! ❌
    "Joining_Date": ["2021-01-15", "2020-06-23", "2022-03-10"]
}
df = pd.DataFrame(data)

print(df.dtypes)

df["Salary"] = df["Salary"].astype(int)

# Fix date column
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])

print(df.dtypes)
```

```python
# Step 1 — check missing values
data = {
    "Name":       ["Rahul", "Priya", None,    "Sneha", "Karan"],
    "Age":        [23,       27,      25,       None,    30],
    "Salary":     [45000,    None,    58000,    40000,   None],
    "Department": ["IT",     "HR",    "IT",     None,    "Sales"]
}
df = pd.DataFrame(data)
print("Before")
print(df)
print(df.isnull().sum())

# Step 2 — fill or drop missing values
df['Age'] = df["Age"].fillna(df["Age"].mean())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Department"] = df["Department"].fillna("Unknown")

# Step 3 — remove duplicates
df = df.drop_duplicates()

# Step 4 — fix data types
df["Salary"] = df["Salary"].astype(int)

# Step 5 — confirm everything is clean
print(df.isnull().sum())   # should all be 0 now!
print(df.shape)             # check rows remaining

print("after")
print(df)
```
