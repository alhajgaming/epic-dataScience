```python
import matplotlib.pyplot as plt
# Zomato daily orders for a week
days = [1, 2, 3, 4, 5, 6, 7]
orders = [1200, 1400, 980, 1700, 2100, 2500, 1900]

plt.plot(days, orders)
plt.show()
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
orders = [1200, 1400, 980, 1700, 2100, 2500, 1900]

plt.plot(days, orders)               # the actual chart

plt.title("Zomato Daily Orders")     # 1. title
plt.xlabel("Day of Week")            # 2. x axis label
plt.ylabel("Number of Orders")       # 3. y axis label
plt.grid(True)                       # 4. gridlines — easier to read
plt.show()                           # 5. display the chart
```

```python
plt.plot(
    days, orders,
    color="blue",        # line colour
    marker="o",          # dot at each data point
    linestyle="--",      # dashed line
    linewidth=2,         # thickness
    markersize=8         # size of dots
)

plt.title("Zomato Daily Orders", fontsize=16)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Orders", fontsize=12)
plt.grid(True)
plt.show()
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
zomato  = [1200, 1400, 980,  1700, 2100, 2500, 1900]
swiggy  = [900,  1100, 1200, 1300, 1800, 2000, 1600]

plt.plot(days, zomato, color="red",  marker="o", label="Zomato")
plt.plot(days, swiggy, color="orange", marker="s", label="Swiggy")

plt.title("Zomato vs Swiggy Daily Orders")
plt.xlabel("Day of Week")
plt.ylabel("Number of Orders")
plt.legend()          # ← shows the label names on chart
plt.grid(True)
plt.show()
```

```python
plt.plot(days, orders, color="blue", marker="o")
plt.title("Daily Orders")
plt.savefig("orders_chart.png")   # saves as image file
plt.show()
```

```python
months = [1, 2, 3, 4, 5, 6]
sales  = [45000, 52000, 48000, 61000, 70000, 65000]
meesho_sales = [30000, 35000, 40000, 38000, 45000, 50000]
plt.plot(months, sales, color="Yellow", label="Flipkart")
plt.plot(months, meesho_sales, color="purple", label="Meesho")
plt.title("Flipkart vs Meesho Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Number Of Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = np.array([10, 25, 38, 42, 55, 60])
result = np.where(data > 40, "High", "Low")
print(result)

import pandas as pd
data = {
    "City":  ["Mumbai", "Delhi", "Mumbai", "Pune", "Delhi", "Mumbai"],
    "Sales": [5000, 3000, 7000, 4000, 6000, 8000]
}
df = pd.DataFrame(data)
result = df.groupby("City")["Sales"].mean().reset_index()
print(result)
```
