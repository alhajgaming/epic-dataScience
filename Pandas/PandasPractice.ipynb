```python
import pandas as pd
```

```python
data = {
    "Name":       ["Rahul","Priya","Arjun","Sneha","Karan","Neha","Amit","Pooja"],
    "Department": ["IT","HR","IT","HR","IT","HR","Sales","Sales"],
    "City":       ["Mumbai","Delhi","Pune","Mumbai","Delhi","Pune","Mumbai","Delhi"],
    "Salary":     [60000, 45000, 75000, 50000, 80000, 48000, 55000, 62000]
}
df = pd.DataFrame(data)
print(df)
```

```python
avg_salary = df.groupby("Department")["Salary"].mean().reset_index()
print(avg_salary)
```

```python
citys = df["City"].value_counts()
print(citys) #Incorrect 
```

```python
result = df.groupby("Department")["Salary"].agg(["sum", "count"]).reset_index()
print(result)
```

```python
data = {
    "Name":     ["Aarav", "Diya", "Rohan", "Ananya", "Vivek", "Ishaan", "Meera", "Kabir"],
    "City":     ["Mumbai", "Delhi", "Mumbai", "Bangalore", "Delhi", "Mumbai", "Bangalore", "Delhi"],
    "Category": ["Electronics", "Clothing", "Electronics", "Clothing", "Electronics", "Clothing", "Electronics", "Clothing"],
    "Sales":    [15000, 8000, 22000, 12000, 18000, 9000, 14000, 11000]
}
df = pd.DataFrame(data)
```

```python
avg_salary = df.groupby("City")["Sales"].mean().reset_index()
print(avg_salary)
```

```python
total_sales = df.groupby("Category")["Sales"].agg(["sum", "count"]).reset_index()
print(total_sales)
```
