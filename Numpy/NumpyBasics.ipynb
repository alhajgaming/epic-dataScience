```python
import numpy as np
import time

size = 1_000_000
list1 = list(range(size))
arr1 = np.array(list1)

start = time.time()
# Python list (loop-based)
list_squares = [x ** 2 for x in list1]
end = time.time()
print(f"Time it takes to multiply in list : {end - start}")
print(f"squares of {list_squares[0:10]}")
# NumPy (vectorized)
start = time.time()
numpy_squares = arr1 ** 2
print(f"Squares of Numpy array : {numpy_squares[0:10]}")
end = time.time()

print(f"Time it takes to multiply in array : {end - start}")
```

# Array indexing and slicing

```python
numpy_array = np.arange(1,30,2)
print(numpy_array)
'''
np.zeros((2,2)) creates 2x2 array with zero values
np.ones((2,2)) creates 2x2 array with ones values
np.eyes(3) creates identity matrix with 3rows and 3 columns
'''
np.linspace(1,10,5) #Creates evenly spaces between 1 to 10 with 5 gap
print(numpy_array.size)
```

```python
reshaped_array = numpy_array.reshape(5,3)
print(reshaped_array)
Flattenarray = reshaped_array.flatten()
print(Flattenarray[0])
```

```python
Flattenarray[0:5]
```

```python
Flattenarray[::-1] #Reverse Array
#We don't need loops for array it can perform vectorized operations
# Slicing return view it doesn't return copy
```

```python
SliceB = Flattenarray[3:7]
SliceB
```

```python
SliceB[0] = 655
#Copy array can also make changes in original array 

print(f"Original Array : {Flattenarray}")
print(f"sliced Array of flatten array : {SliceB}")
```

# slice is view and it affects original memory location 
- that's wny we have to use the .copy() method to create copy of an array in numpy

```python
b = Flattenarray[0:5].copy()
b[0] = 67


print(f"Orignal Array: {Flattenarray}")
print(f"sliced view array : {SliceB}")
print(f"Copy array of original : {b}")
```
