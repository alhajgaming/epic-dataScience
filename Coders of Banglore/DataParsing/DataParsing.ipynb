```python
with open('initialdata.txt', encoding='utf-8') as f:
    data = f.read()

print(data)
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunks):
    # try:
        chunk = chunks.strip()
        seprated_chunk  = chunk.split('\n')
        username = seprated_chunk[0]
        number_of_posts = seprated_chunk[1]
        number_of_followers = seprated_chunk[2]
        number_of_following = seprated_chunk[3]
        name = seprated_chunk[4]
        if len(seprated_chunk) > 5:
            Type_of_page = seprated_chunk[5]
            bio = '\n'.join(seprated_chunk[6:])
        else:
            Type_of_page = "Unknown"
            bio = "none"
        # print(  username, number_of_posts, number_of_followers, number_of_following, name, Type_of_page, bio, sep='\n' )
        return {
            'username': username,
            'number_of_posts': number_of_posts,
            'number_of_followers': number_of_followers,
            'number_of_following': number_of_following,
            'name': name,
            'Type_of_page': Type_of_page,
            'bio': bio
        }
    # except Exception as e:
    #     print(f"Error parsing chunk: {e}")
    #     print(f"Chunk content: {chunk}")


# parse_chunk(chunk[0])
```

```python
all_chunks = []

for chunk in chunks:          # ← 'chunks' not 'chunk'
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)

print(all_chunks)
```

```python
import json
json_data = json.dumps(all_chunks, indent=4, ensure_ascii=False)
with open('parsed_data.json', 'w', encoding='utf-8') as f:
    f.write(json_data)
```
