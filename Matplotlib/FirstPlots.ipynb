```python
import matplotlib.pyplot as plt
print("imported")
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs, label="Runs")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years, kohli, 'g^:', label="Kohli")
plt.plot(years, sehwag, 'ro-.', label="Virender Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
plt.plot(years, kohli, color='orange', linestyle='--',marker=',', label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.',marker='.', label="Sehwag")
plt.plot(years, runs, color='blue',linestyle='dashed',marker='^', label="Tendulkar")
plt.legend()
```

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
