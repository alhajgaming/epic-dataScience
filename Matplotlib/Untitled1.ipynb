```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]  # hours in a day

colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(6,6))
plt.pie(time_spent, labels=activities, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("After School Activities")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]

labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']

plt.figure(figsize=(10,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping,
              labels=labels, colors=colors, alpha=0.8)

plt.legend(loc='upper right')  # Location of the legend
plt.title('Weekly Activity Tracker')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```

```python
#Personal routine analysis
import random
Days = [1,2,3,4,5,6,7]

sleeping = [7,6,6,7,5,8,8]
workout = [2,1,2,2,2,2,2]
office = [8,8,8,8,8,8,8]
coding = [1,1.5,2,1,1,1,1]
study = [2,2,2,1,2,2,2]

labels =['Sleeping', 'Workout', 'Office Hours','Coding', 'Study']
colors = [
    "#" + ''.join(random.choices('0123456789ABCDEF', k=6))
    for _ in range(len(labels))
]
plt.figure(figsize=(10,6))
plt.stackplot(Days, sleeping, workout, office, coding, study, labels=labels, colors=colors, alpha=0.8)
plt.title("Weekly time analysis")
plt.xlabel("Days")
plt.ylabel("Activities")
plt.legend()
plt.grid()
plt.show()
```
