```python
import matplotlib.pyplot as plt

print("Mission passed")
plt.style.use("ggplot")
# plt.style.available
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '--'}
)
plt.show()
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice
 
plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=140
)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Career Run Share")
plt.show()
```

```python

platforms = ["WhatsApp", "Instagram", "Facebook", "TikTok", "X"]
colors = ["green", "#F77737", "blue", "black", "skyblue"]
users = [3000, 2400, 3100, 1800, 650]

plt.barh(platforms, users, color=colors)

plt.title("Global Social Media Users (Millions)")
plt.xlabel("Users (Millions)")
plt.ylabel("Platforms")

plt.show()
```
