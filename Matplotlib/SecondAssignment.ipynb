```python
import numpy as np
import matplotlib.pyplot as plt
print("Libraries are here")
```

```python
# Selected years
years = np.array([1, 2, 3, 4, 5])

# Runs scored in those years (hypothetical data)
sachin = np.array([850, 920, 980, 1050, 1100])
kohli  = np.array([780, 890, 950, 1020, 1080])
sehwag = np.array([700, 820, 910, 970, 1030])

width = 0.25
years = np.array([1990,1995,2000,2005,2010])
x = np.arange(len(years))
```

```python
plt.style.use("ggplot")
plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")
plt.title("Runs comparison:")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Runs scored")
plt.xticks(x, years)
plt.grid()
```
