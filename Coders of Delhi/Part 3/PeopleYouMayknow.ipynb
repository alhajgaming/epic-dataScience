```python
import json

def load_data(filename):
    with open(filename, 'r') as f:
        return json.load(f)
    
def People_You_May_know(user_id, data):
    user_freinds = {}
    for user in data['users']:
        user_freinds[user['id']] = set(user['friends'])

    if user_id not in user_freinds:
        return[]
    direct_friends = user_freinds[user_id]
    suggestion = {}
    for friend in direct_friends:
        for mutual in user_freinds[friend]:
            if mutual != user_id and mutual not in direct_friends:
                suggestion[mutual] = suggestion.get(mutual, 0) + 1
            
    sorted_suggestions = sorted(suggestion.items(), key=lambda x: x[1], reverse=True)
    return [user_id for user_id, _ in sorted_suggestions]


data = load_data("MassiveData.json")
user_id = 1
recc = People_You_May_know(user_id, data)
print(f"People you may know for user {user_id}: {recc}")
```
